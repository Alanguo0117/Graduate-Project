```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([0, 1, 2])
y = np.array([0.98495, 0.98865, 0.9887])
y_max = [0.9892, 0.9900, 0.9910]
y_min = [0.9801, 0.9873, 0.9861]
lower_err = np.array(y) - np.array(y_min)
upper_err = np.array(y_max) - np.array(y)
asy_err = [lower_err, upper_err]
#print(lower_err, upper_err, asy_err)

plt.errorbar(x, y, yerr=asy_err, fmt='-o')
plt.plot(y_max,'-o')
plt.plot(y_min,'o')
plt.xticks(np.arange(0, 3, 1))
plt.yticks(np.arange(0.9800, 0.9920, 0.001))
plt.ylabel('Accuracy')
plt.xlabel('Generation')
plt.title('Accuracy of Popluation')
plt.legend(['max', 'min','average'], loc = 'lower right')
plt.savefig('./3gen_test.png')
```
